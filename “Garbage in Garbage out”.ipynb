                                            
                                           Exploring the data
                                           
Deciding how to sort the data. 

Considering that almost one quater of each column is missing its values. I was not expecting the measures of central tensdancy to be so similar. The mean of each column and the median Q2 of each column do not deviate drastically from one and other. As these two measures of central tendancy are similar it is a sign that the data has a symmetrical distribution. The 68 - 95 - 99.7% rule, which means that 68% of the sample should be within one SD of the mean, 95% within 2 SD and 99,7% within 3 SD. 

According to the information below, plastic makes up a relatively low percentage of the worlds average waste (11.69%). This is a point of interest to me as it feels counter intuitive. According to the data the world produces a similar amount of plastic and garden waste - (11.41 mena%) On the other hand, agricultural waste, waste that one might presume to be biodegradable, exhibits one of the largest average percentages.

```python
# Indexing and selecting data below the median gdp. Making pd df for countries with low gdp and sorting in order of gdp. 
CO_LOW = CODF.loc[CODF['gdp'] <= 13465.885250]
CO_LOW = CO_LOW.sort_values(by='gdp', ascending=False)
CO_LOW_GDP = pd.DataFrame(CO_LOW)
```

```python
CountryLGCO = CODF.loc[[66,100,103,135,146,170,192,197,216]]
CountryLGCO = pd.DataFrame(CountryLGCO)
Col_name = CountryLGCO.head()
```

```python
Imports2 = Imports.loc[['FSM','KEN','KIR','MRT','NPL','SLB','TLS','TUV','ZWE']]
Imports2 = pd.DataFrame(Imports2)
```

```python
Export2 = Exports.loc[['FSM','KEN','KIR','MRT','NPL','SLB','TLS','TUV','ZWE']]
Export2 = pd.DataFrame(Export2)
```

```python
## X = Imports. Y = Exports

merge_df = pd.merge(Imports2, Export2, on='economy', how='left').fillna(0)
merge_df.to_csv('merged.csv', index=False)
print(merge_df)
```

```python
#### Mergeing data from imports and exports of countries with low GDP and above average plastic, 
### with all garbage related variables for those said countries.  

merge_df_ALL = pd.merge(merge_df, CountryLGCO, on='economy', how='left').fillna(0)
merge_df_ALL.to_csv('merge_df_ALL', index=False)
merge_df_ALL = pd.DataFrame(merge_df_ALL)
Col_name = merge_df_ALL.head()
print(Col_name)
```
